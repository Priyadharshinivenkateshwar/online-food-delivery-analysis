```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned_food_analysis.csv")
df.head()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

df['Order_Time'] = pd.to_datetime(
    df['Order_Time'], format='%H:%M:%S', errors='coerce'
).dt.time
```

```python
df['Order_Day_Type'] = df['Order_Date'].dt.dayofweek.apply(
    lambda x: 'Weekend' if x >= 5 else 'Weekday')
```

```python
df['Peak_Hour_Flag'] = df['Peak_Hour'].apply(
    lambda x: 'Peak' if str(x).lower() == 'peak' else 'Non-Peak')
```

```python
df['Profit_Margin_Pct'] = np.where(
    df['Final_Amount'] > 0,
    (df['Profit_Margin'] / df['Final_Amount']) * 100,0)
```

```python
df['Delivery_Performance'] = pd.cut(
    df['Delivery_Time_Min'],
    bins=[0, 30, 45, 60, np.inf],
    labels=['Fast', 'On-Time', 'Delayed', 'Very Delayed'])
```

```python
df['Customer_Age_Group'] = pd.cut(
    df['Customer_Age'],
    bins=[0, 18, 25, 35, 45, 60, np.inf],
    labels=['<18', '18-25', '26-35', '36-45', '46-60', '60+'])
```

```python
df[['Order_Day_Type', 'Order_Day_Type', 'Peak_Hour_Flag', 'Profit_Margin_Pct',
     'Delivery_Performance','Customer_Age_Group']].head()
```

```python
df.to_csv("../data/featured_engineered_food_analysis.csv",index=False)
```
